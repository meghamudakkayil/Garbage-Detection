```python
import zipfile, os

zip_path = "/content/drive/MyDrive/garbage_Project/small.zip"
extract_path = "/content/small"


with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
print("Done")

for split in ["train", "valid", "test"]:
    img_count = len(os.listdir(f"{extract_path}/images/{split}"))
    print(f"{split}: {img_count} images")
```

```python
import yaml

yaml_path = "/content/small/data.yaml"

with open(yaml_path, "r") as f:
    data = yaml.safe_load(f)

data["path"] = "/content/small"
data["train"] = "images/train"
data["val"] = "images/valid"
data["test"] = "images/test"

with open(yaml_path, "w") as f:
    yaml.dump(data, f, default_flow_style=False)

print("Fixed")
print(open(yaml_path).read())
```

```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")

results = model.train(
    data="/content/small/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    patience=10,
    device=0,
    project="/content/runs",
    name="garbage_detector_v1",
    exist_ok=True
)

print("Training complete")
```
